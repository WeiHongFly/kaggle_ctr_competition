### 逐列进行检测：空值检测，unique count,label的分布是否均衡

```python
for col in train_df_sample.columns.values:
    the_df=pd.read_csv(train_dir,usecols=[col])
    print("col:{}  unique-count:{}, null-count:{}, all-rows:{}".format(col,the_df[col].unique().shape[0],the_df[col].isna().sum(),the_df.shape[0]))
    
    df1=the_df[col].value_counts().to_frame()
    df1["rate"]=df1[col]/df1[col].sum(axis=0)
    df1.sort_values(by=[col],ascending=False,inplace=True)
    display(df1)    
    print("\n")
```
